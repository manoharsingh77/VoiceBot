```python
# -------------------------------------------------------
# 1. INSTALL DEPENDENCIES
# -------------------------------------------------------
!pip install flask pyngrok

# -------------------------------------------------------
# 2. CONFIGURATION (ENTER YOUR TOKEN HERE)
# -------------------------------------------------------
# Go to https://dashboard.ngrok.com/get-started/your-authtoken to get your token
NGROK_AUTH_TOKEN = "35si0TbOw6w83WE3ViqruwuhsaH_6hH3cZxn4vi5zVxcyRSvY"

# -------------------------------------------------------
# 3. SETUP FOLDERS & FILES
# -------------------------------------------------------
import os

# Create templates folder
if not os.path.exists('templates'):
    os.makedirs('templates')

# Write dashboard.html
dashboard_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Voice Bot Analytics</title>
    <style>
        body { font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; margin: 0; padding: 20px; background-color: #f4f4f9; }
        h1 { color: #333; }
        .metrics { display: flex; gap: 20px; margin-bottom: 30px; }
        .card { background: white; padding: 20px; border-radius: 8px; box-shadow: 0 2px 5px rgba(0,0,0,0.1); flex: 1; text-align: center; }
        .card h2 { margin: 0; font-size: 2.5em; color: #007bff; }
        .card p { margin: 5px 0 0; color: #666; }
        table { width: 100%; border-collapse: collapse; background: white; border-radius: 8px; overflow: hidden; box-shadow: 0 2px 5px rgba(0,0,0,0.1); }
        th, td { padding: 15px; text-align: left; border-bottom: 1px solid #eee; }
        th { background-color: #007bff; color: white; }
        tr:hover { background-color: #f1f1f1; }
    </style>
</head>
<body>
    <h1>ðŸ¤– Voice Bot Analytics Dashboard</h1>

    <div class="metrics">
        <div class="card">
            <h2>{{ total }}</h2>
            <p>Total Interactions</p>
        </div>
        <div class="card">
            <h2>{{ rate }}%</h2>
            <p>Success Rate</p>
        </div>
        <div class="card">
            <h2>0.5s</h2>
            <p>Avg Response Time</p>
        </div>
    </div>

    <h2>Recent Call Logs</h2>
    <table>
        <thead>
            <tr>
                <th>Timestamp</th>
                <th>Intent Tag</th>
                <th>User Input Type</th>
                <th>Status</th>
            </tr>
        </thead>
        <tbody>
            {% for log in logs %}
            <tr>
                <td>{{ log.timestamp }}</td>
                <td>{{ log.intent }}</td>
                <td>{{ log.query }}</td>
                <td>{{ log.status }}</td>
            </tr>
            {% endfor %}
        </tbody>
    </table>
</body>
</html>
"""
with open('templates/dashboard.html', 'w') as f:
    f.write(dashboard_code)

# Write mock_db.json
mock_db_code = """
{
    "accounts": {
        "12345": { "name": "John Doe", "balance": 4500.50, "type": "Savings" },
        "67890": { "name": "Jane Smith", "balance": 120.75, "type": "Checking" },
        "11223": { "name": "Alice Johnson", "balance": 9800.00, "type": "Investment" }
    }
}
"""
with open('mock_db.json', 'w') as f:
    f.write(mock_db_code)

# -------------------------------------------------------
# 4. FLASK APP CODE
# -------------------------------------------------------
import json
import datetime
import threading
from flask import Flask, request, jsonify, render_template
from pyngrok import ngrok

app = Flask(__name__)
DB_FILE = 'mock_db.json'
LOG_FILE = 'call_logs.json'

def load_db():
    if not os.path.exists(DB_FILE): return {"accounts": {}}
    with open(DB_FILE, 'r') as f: return json.load(f)

def load_logs():
    if not os.path.exists(LOG_FILE): return []
    with open(LOG_FILE, 'r') as f: return json.load(f)

def save_log(intent_name, user_query, status="Success"):
    logs = load_logs()
    entry = {
        "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "intent": intent_name,
        "query": user_query,
        "status": status
    }
    logs.append(entry)
    with open(LOG_FILE, 'w') as f: json.dump(logs, f, indent=4)

def get_balance(account_number):
    db = load_db()
    account = db.get("accounts", {}).get(account_number)
    if account: return f"Your current balance is ${account['balance']}."
    return None

@app.route('/')
def home():
    return "Voice Bot Webhook Service is Running on Colab!"

@app.route('/webhook', methods=['POST'])
def webhook():
    req = request.get_json(silent=True, force=True)
    tag = req["fulfillmentInfo"]["tag"]
    parameters = req["sessionInfo"]["parameters"]
    save_log(tag, "Voice/Text Input")

    response_text = ""
    if tag == "check_balance":
        acc_num = str(int(parameters.get("account_number", 0)))
        balance_msg = get_balance(acc_num)
        response_text = balance_msg if balance_msg else f"I could not find account {acc_num}."
    else:
        response_text = f"No logic for tag: {tag}"

    return jsonify({
        "fulfillment_response": {
            "messages": [{"text": {"text": [response_text]}}]
        }
    })

@app.route('/dashboard')
def dashboard():
    logs = load_logs()
    return render_template('dashboard.html', logs=logs, total=len(logs), rate=100)

# -------------------------------------------------------
# 5. START SERVER & NGROK
# -------------------------------------------------------
# Authenticate ngrok
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Open a ngrok tunnel to the HTTP server
public_url = ngrok.connect(5000).public_url
print("---------------------------------------------------------")
print(f"ðŸš€ COPY THIS URL FOR DIALOGFLOW WEBHOOK: {public_url}/webhook")
print(f"ðŸ“Š OPEN DASHBOARD HERE: {public_url}/dashboard")
print("---------------------------------------------------------")

# Run the app
if __name__ == '__main__':
    app.run(port=5000)
```
